### This combines the csv files generated from Transformer batch processing and merges them with all of the other NLP metrics previously calculated. This merged df is output from this file and used as an input for Classifier ML model training 

Fed Data

```python
FED_df = pd.concat(map(pd.read_csv, ['Batch_transformer_files/sentiment_minutes_08_10.csv', 'Batch_transformer_files/sentiment_minutes_10_13.csv','Batch_transformer_files/sentiment_minutes_14_18.csv','Batch_transformer_files/sentiment_minutes2.csv','Batch_transformer_files/sentiment_minutes_22_23.csv']))
USE_scores = pd.read_csv('Batch_transformer_files/use_scores_07_20.csv')
FED_df = FED_df.drop(columns=['Unnamed: 0'])
FED_idx = FED_df['index']
FED_df.reset_index(inplace=True)
FED_df =FED_df.drop(['level_0'],axis=1)
FED_df.to_csv('transformer_sentiment.csv')
FED_df
```

Combine all the Other NLP Data

```python
NLP_data = pd.read_csv('Batch_transformer_files/NLP_output_07_19.csv') # LDA, Cosine Sim, d2V
NER_data = pd.read_csv('Batch_transformer_files/NLP_FOMC_people_07_19.csv') # Named entities 
NLP_data = NLP_data.rename(columns ={'Unnamed: 0': 'index'})
NLP_data['USE_Score'] = USE_scores
NER_data = NER_data.rename(columns ={'Unnamed: 0': 'index'})

```

# NLP DF Containing all of the NLP Metrics we calculated for the project

Merging the previous NLP data to the Transformer Data which had to be processed separately

```python
# Merge dfs
temp = pd.merge(FED_df,NLP_data, how='left', on = 'index', suffixes = ('_left', '_right'))
df=temp
df
```
